1. What are data structures, and why are they important?
Data structures are ways to store and organize data efficiently. They make programs faster and more optimized.

2. Difference between mutable and immutable data types

Mutable data types can be changed after they are created.

Examples: list, dictionary, set

You can add, remove, or modify their elements.

Immutable data types cannot be changed once created.

Examples: int, float, string, tuple

Any change creates a new object instead of modifying the existing one.

3. Differences between lists and tuples.

Lists are mutable, meaning their elements can be modified.

Tuples are immutable, meaning their elements cannot be changed.

Lists use square brackets [ ], while tuples use parentheses ( ).
Lists are generally used for data that needs to change, while tuples are used for fixed data.

4. How dictionaries store data
